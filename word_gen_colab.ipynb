```python
print("tępy chuju")
print("pojebana kurwo")
import fasttext
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# --- 1. Load pre-trained FastText embeddings (English example) ---
# Use the official wiki-news-300d-1M.vec or similar
# We'll use gensim KeyedVectors for convenience
from gensim.models import KeyedVectors

# Load FastText word vectors (only load the .vec file)
# path = "wiki-news-300d-1M.vec"
# Make sure to download FastText 300d embeddings beforehand
print("Loading word vectors...")
wv = KeyedVectors.load_word2vec_format("wiki-news-300d-1M.vec", binary=False)
print("Word vectors loaded successfully.")

# --- 2. Pick words A, B, C ---
word_A = "king"
word_B = "queen"
word_C = "man"
print(f"Selected words: A='{word_A}', B='{word_B}', C='{word_C}'")

# --- 3. Compute vector difference ---
v_A = wv[word_A]
v_B = wv[word_B]
X_diff = v_B - v_A  # Semantic transformation
print(f"Vector difference (B-A) computed. Shape: {X_diff.shape}")

# --- 4. Apply difference to C to get candidate vector D ---
v_C = wv[word_C]
v_D_candidate = v_C + X_diff
print(f"Candidate vector D computed. Shape: {v_D_candidate.shape}")

# --- 5. Find nearest neighbor to v_D_candidate using gensim's most_similar ---
# Gensim's most_similar method is optimized for this
# We need to provide positive and negative words, or a vector directly.
# The analogy is C + (B - A), which can be written as positive=[B, C], negative=[A]
# Or, directly use the candidate vector v_D_candidate

# Using the vector directly:
print("Searching for nearest neighbor...")
similar_words = wv.most_similar(positive=[v_D_candidate], topn=1)

word_D = similar_words[0][0]
similarity = similar_words[0][1]
print("Nearest neighbor search complete.")

print(f"Candidate D for {word_C} + ({word_B}-{word_A}): {word_D}")
print(f"Cosine similarity: {similarity:.4f}")

# Alternative using positive/negative words directly for analogy (often more robust):
# most_similar_analogy = wv.most_similar(positive=[word_B, word_C], negative=[word_A], topn=1)
# print(f"Candidate D (using analogy words) for {word_C} + ({word_B}-{word_A}): {most_similar_analogy[0][0]}")
# print(f"Cosine similarity (using analogy words): {most_similar_analogy[0][1]:.4f}")
```
